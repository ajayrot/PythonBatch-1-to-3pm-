# Matplotlib - Markers


```python
import matplotlib.pyplot as plt
import numpy as np

y_axis = np.array([13,8,1,10])

plt.plot(y_axis,marker = 'D')
plt.show()
```

```python
#marker with plot color
import matplotlib.pyplot as plt
import numpy as np
x_axis = np.array([13,1,6,8])
plt.plot(x_axis, '*:k',ms = 30,mec ='r',mfc = 'g')#ms set size of the marker
plt.show() #mec set color to edge of marker
#mfc set color inside the marker
```

```python
# Linesytle - ls
#solid
# dotted
#dashed
#dashdot
import matplotlib.pyplot as plt
import numpy as np
y_axis = np.array([3,8,6,1])
plt.plot(y_axis,ls = "dashdot", color = 'r',linewidth= '10')
plt.show()
```

```python
#multiple lines
x1 = np.array([1,3,5,7])
x2 = np.array([3,6,8,11])
y1= np.array([3,8,5,1])
y2 =np.array([6,2,7,9])
plt.plot(x1,ls = "solid", color = 'y',linewidth= '3')
plt.plot(x2,ls = "dashdot", color = 'b',linewidth= '2')
plt.plot(y1,ls = "dashdot", color = 'r',linewidth= '5')
plt.plot(y2,ls = 'dotted' , color = 'g' ,linewidth = '7')
plt.show()
```

```python
x1 = np.array([1,3,5,7])
x2 = np.array([3,6,8,11])
y1= np.array([3,8,5,1])
y2 =np.array([6,2,7,9])
plt.title("Graph")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.plot(x1,ls = "solid", color = 'y',linewidth= '3')
plt.plot(x2,ls = "dashdot", color = 'b',linewidth= '2')
plt.plot(y1,ls = "dashdot", color = 'r',linewidth= '5')
plt.plot(y2,ls = 'dotted' , color = 'g' ,linewidth = '7')
plt.grid()
plt.show()
```

```python
x1 = np.array([1,3,5,7])
x2 = np.array([3,6,8,11])
plt.bar(x1,x2)
plt.title("Graph")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.grid()
plt.show()
```

```python
x1 = np.array([1,3,5,7])
x2 = np.array([3,6,8,11])
plt.barh(x1,x2)
plt.title("Graph")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.grid()
plt.show()
```
